```python

import pandas as pd
import matplotlib as plt
df = pd.read_csv("foodstruct_nutritional_facts.txt",sep='\t', header=None, names=['Food Name','Category Name','Calories','Fats','Fiber'])
df
#changing data types 
df['Fats'] = pd.to_numeric(df['Fats'], errors='coerce')
df['Calories'] = pd.to_numeric(df['Calories'], errors='coerce')
df['Fiber'] = pd.to_numeric(df['Fiber'], errors='coerce')
print(df.dtypes)
#Calculating Mean 
df[['Fats','Calories','Fiber']].mean()
#Calculating mean,variance and Standard deviation for only Calories
mean = df['Calories'].mean()
var = df['Calories'].var()
std = df['Calories'].std()
print(f"Mean ={mean}\nVariance = {var}\nStandard Deviation = {std}")
#Histogram
df['Calories'].hist(bins=15, figsize=(10,6))
```
